```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import IPython.display
from IPython.display import Image

```

```python

training_set = (r'/Users/henrywillett/Downloads/CHY2101 JGK/CHY2301/CHY3201/Assignments 2022/Machine Learning/Machine Learning in Science/machine-learning-in-science-ii-2023/training_data/training_data')
test_set = (r'/Users/henrywillett/Downloads/CHY2101 JGK/CHY2301/CHY3201/Assignments 2022/Machine Learning/Machine Learning in Science/machine-learning-in-science-ii-2023/test_data/test_data')
```

```python
model = Sequential()

model.add(Conv2D(filters=16,kernel_size=2,padding='same',activation='relu',input_shape=(320,240,3)))
model.add(MaxPooling2D(pool_size=2))

model.add(Conv2D(filters=32,kernel_size=2,padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64,kernel_size=2,padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.compile(loss='categorical_cross_entropy',optimizer='adam',metrics=['accuracy'])

model.add(Flatten())

#Fully connected layer

model.add(Dense(500,activation='relu'))
model.add(Dense(2,activation='softmax'))

model.fit_generator(training_set,validation_data=test_set,epochs=50,steps_per_epoch=len(training_set), validation_steps=len(test_set))

```
